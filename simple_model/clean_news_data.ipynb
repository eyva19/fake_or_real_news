## Step 1 - Clean the data

```python
#check count of all rows
news_df.describe()
```

```python
#drop NAs from all rows
news_df = news_df.dropna()
```

```python
news_df.head()
```

```python
#recheck all rows
news_df.describe()
```

```python
#see datatype
news_df.dtypes
```

```python
# Changing the data type of 'id' to float

news_df['id'] = pd.to_numeric(news_df['id'], errors='coerce')

# Using .loc to modify the 'id' column
# news_df_cleaned.loc[:, 'id'] = pd.to_numeric(news_df_cleaned['id'], errors='coerce')
```

```python
news_df.dtypes
```

```python
# Changing the data type of 'title' 'text' 'label' to string

news_df['title'] = news_df['title'].astype(str)
news_df['text'] = news_df['text'].astype(str)
news_df['label'] = news_df['label'].astype(str)
```

```python
news_df = news_df.astype(dtype={'title': 'string', 'text': 'string', 'label': 'string'})
```

```python
news_df.head()
```

```python
news_df.dtypes
```

```python
#check to see that the values are only fake of real ONLY
news_df['label'].value_counts()
```

```python
# Drop all rows that are not "fake" or "real" from column "label". convert it upper string first so it is not case sensitive 
news_df['label'] = news_df['label'].str.upper()
cleaned_news_df = news_df[news_df['label'].isin(['FAKE', 'REAL'])]
cleaned_news_df.head()
```

```python
#check again 
cleaned_news_df['label'].value_counts()
```

```python
# save file 
cleaned_news_df.to_csv('Resources/Data/cleaned_news_df.csv', index=False)
```
